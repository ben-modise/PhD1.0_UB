```python
py_table1 = pd.DataFrame(columns = table1.columns)
```

```python
#run problems and record to table
n = 0
for F in problems:
    n += 1
    for dim in dimensions:
        x0 = np.ones([dim,1])
        Pn = str(n) + '(' + str(dim) + ')'
        x_star, iters, Fevals, norm_F = DMBHP(F, x0)
        row = [Pn, iters, Fevals, norm_F]
        py_table1.loc[len(py_table1)] = row
        
```
